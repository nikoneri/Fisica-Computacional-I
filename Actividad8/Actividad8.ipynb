```python
from scipy.integrate import odeint
import numpy as np

def vectorfield(w, t, p):
    """
    Arguments:
        w :  vector of the state variables:
                  w = [x,y]
        t :  time
        p :  vector of the parameters:
                  p = [mu]
    """
    x, y = w

    # Create f = (x',y'):
    f = [y,
         (p*(1.0 - x**2.0) * y) - x]
    return f
```

```python
#Lista de datos generados por el ODE solver para el parametro
# ****************** mu=0.1 *******************

mu = 0.1
# Initial conditions
# x is the initial displacement and y is the initial velocity
x = 2.0
y = -3.0

stoptime = 100
numpoints = 3000

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
w0 = [x, y]

# Call the ODE solver.
wsol = odeint(vectorfield, w0, t, args=(mu,))

with open('vanderpol_0', 'w') as f:
    # Print & save the solution.
    for t1, w1 in zip(t, wsol):
        print (t1, w1[0], w1[1], file=f)
```

```python
#Ejemplo del Oscilador de Van der Pol Forzado

import numpy as np
def vanpolforced(w, t, p):
    """
    Defines the differential equation for the Van der Pol oscillation system.

    Arguments:
        w :  vector of the state variables:
                  w = [x,y]
        t :  time
        p :  vector of the parameters:
                  p = [mu]
    """
    x, y = w
    
    A, om, mu = p

    # Create f = (x',y'):
    f = [y,
         mu*(1.0 - x**2.0) * y - x + A*np.sin(om*t)]
    return f


# Use ODEINT to solve the differential equations defined by vanpolforced
from scipy.integrate import odeint
from numpy import *

# Parameter values
# Non-linear Damping parameter:
mu = 8.53
#Amplitude of forcing
A = 1.2
#Angular frecuency of forcing
omega = 2.0*pi/10.0


# Initial conditions
# x is the initial displacement and y is the initial velocity
x = 1.0
y = -10.0

stoptime = 1000
numpoints = 5000

# Create the time samples for the output of the ODE solver.
# I use a large number of points, only because I want to make
# a plot of the solution that looks nice.
t = [stoptime * float(i) / (numpoints - 1) for i in range(numpoints)]

# Pack up the parameters and initial conditions:
w0 = [x, y]

p=[A,omega,mu]

# Call the ODE solver.
wsol = odeint(vanpolforced, w0, t, args=(p,))

with open('vanderpol_forced.dat', 'w') as f:
    # Print & save the solution.
    for t1, w1 in zip(t, wsol):
        print (t1, w1[0], w1[1], file=f)
```

```python
# Plot the solution that was generated

from numpy import loadtxt
from pylab import figure, plot, ylabel, xlabel, grid, hold, legend, title, savefig, ylim, xlim
from matplotlib.font_manager import FontProperties
%matplotlib inline

t, x, y = loadtxt('vanderpol_forced.dat', unpack=True)

figure(1, figsize=(12, 2))

ylabel('$x_1$')
xlabel('t')
grid(True)
#hold(True)
lw = 2

ylim(-2.5, 2.5)
xlim(300,600)

plot(t, x, 'b', linewidth=lw)

savefig('Figura4.png', dpi=100)
```
