```python
# Cargar a la memoria de trabajo las bibliotecas: Pandas (manejo de datos, 
# Numpy (numerical python) y la biblioteca de gráficas Matplotlib
# Se asignan nombres cortos.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#
# Usar "Shift+Enter" para procesar la información de la celda
#
```

```python
df0 = pd.read_csv('Iguala.txt', skiprows=4, sep='\s+')
#df0 es la variable que guarda el archivo
#read_csv lee datos con elementos separados por mas de un espacio
```

```python
# Lee los primeros 5 renglones del archivo
df0.head(50)
# "Shift+Enter"
```

```python
# Dar estructura de datos (DataFrame)
#It refers to "tabular" data, a data structure using rows and columns.
df = pd.DataFrame(df0)
```

```python
# Ver los tipos de datos que Pandas ha reconocido al leer
df.dtypes
```

```python
# Combinar las columnas "DD/MM/AAAA" con "HH:MM" y convertirla a variable de tiempo
# Se crea una nueva columna "Fecha" al final con formato de tiempo.
# Eliminamos las dos primeras columnas que ya no necesitaremos
df['FECHA'] = pd.to_datetime(df.apply(lambda x: x['FECHA'] + ' ' + x['HORAS'], 1))
df = df.drop(['FECHA', 'HORAS'], 1)
```

```python
df.head()
```

```python
# Realiza un análisis exploratorio de datos
df.describe()
#Obtienes datos como el promedio, desviacion estandar, min, max, etc.
```

```python
# Selecciona los renglones con Temperatura > 24ºC y < 25ºC
df_tmp = df[df.TEMP > 24] 
df_select = df_tmp[df_tmp.TEMP < 25]
df_select
```

```python
# Calcula el promedio de las columnas, excepto en la FECHA (que no tendría sentido)
df.mean()
```

```python
# Calcula el promedio de las Temperaturas
df.TEMP.mean()
```

```python
# Gráfica de la rapidez de los vientos (m/s) 
plt.figure(); df.VELS.plot(); plt.legend(loc='best')
plt.title("Variación de la Rapidez de los Vientos")
plt.ylabel("Rapidez (m/s)")
plt.grid(True)
plt.show()

```

```python
# Gráfica de Temperatura y Humedad Relativa
df1 = df[['TEMP','HR']]
plt.figure(); df1.plot(); plt.legend(loc='best')
plt.title("Variación de la Temperatura y la Humedad Relativa")
plt.ylabel("Temp ºC /(%) HR")
plt.grid(True)
plt.show()
```

```python
plt.plot_date(x=df.FECHA, y=df.TEMP, fmt="b-")
plt.title("Variación de la Temperatura")
plt.ylabel("Temp ºC")
plt.grid(True)
plt.show()
```

```python
# Gráfica de la rapidez de los vientos y rapidez de las ráfagas
df1 = df[['VELS','VELR']]
plt.figure(); df1.plot(); plt.legend(loc='best')
plt.title("Variación rapidez de los vientos/rapidez de las ráfagas")
plt.ylabel("Rapidez de los vientos (m/s) /Rapidez de las ráfagas")
plt.grid(True)
plt.show()
```

```python
# Lee los primeros 5 renglones del archivo
df2.head(50)
# "Shift+Enter"
```

```python
# Dar estructura de datos (DataFrame)
#It refers to "tabular" data, a data structure using rows and columns.
df = pd.DataFrame(df2)
```

```python
plt.plot_date(x=df.HORAS, y=df.VELS, fmt="b-")
plt.plot_date(x=df.HORAS, y=df.VELR, fmt="b-")
plt.xticks(rotation=90)
plt.title("Variación de la rapidez de los vientos/rapidez de las ráfagas")
plt.ylabel("Rapidez vientos(m/s)/ Rapidez ráfagas(m/s)")
plt.xlabel("Tiempo (horas)")
plt.grid(True)
plt.show()
```

```python
plt.plot_date(x=df.HORAS, y=df.DIRS, fmt="b-")
plt.xticks(rotation=90)
plt.title("Dirección de los vientos")
plt.ylabel("Dir. Vientos (grados)")
plt.xlabel("Tiempo (horas)")
plt.grid(True)
plt.show()
```

```python
plt.plot_date(x=df.HORAS, y=df.RADSOL, fmt="b-")
plt.xticks(rotation=90)
plt.title("RADIACION SOLAR")
plt.ylabel("Rad. Solar (W/m^2)")
plt.xlabel("Tiempo (horas)")
plt.grid(True)
plt.show()
```
