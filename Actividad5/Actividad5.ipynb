```python
#Importar bibliotecas:
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#Leer archivo 00z
df00z= pd.read_csv("df2017CAPE_PW_00Z.csv", header=None, names=["Date" , "CAPE", "PW"])
#Convertir la columna CAPE de objeto a número
df00z.CAPE=pd.to_numeric(df00z.CAPE, errors="coerce")  
df00z.head()                                                                
                                                                
```

```python
#Leer archivo 12z
df12z= pd.read_csv("df2017CAPE_PW_12Z.csv", header=None, names=["Date" , "CAPE", "PW"])
#Convertir la columna CAPE de objeto a número
df12z.CAPE=pd.to_numeric(df12z.CAPE, errors="coerce")  
df12z.head()   
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate' para 00z
df00z['Ndate'] = pd.to_datetime(df00z['Date'], format='%d %m %Y')
df00z['month'] = df00z['Ndate'].dt.month
df00z.head()
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate' para 12z
df12z['Ndate'] = pd.to_datetime(df12z['Date'], format='%d %m %Y')
df12z['month'] = df12z['Ndate'].dt.month
df12z.head()
```

```python
# graficar Boxplots por mes (df00z)
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="CAPE", data=df00z)
plt.show()
```

```python
# graficar Boxplots por mes (df00z)
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="PW", data=df00z)
plt.show()
```

```python
# graficar Boxplots por mes (df12z)
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="CAPE", data=df12z)
plt.show()
```

```python
# graficar Boxplots por mes (df12z)
# Biblioteca Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="PW", data=df12z)
plt.show()
```

```python
import seaborn as sns   #Para df00z
sns.set(style="darkgrid", color_codes=True)

g = sns.jointplot("CAPE", "PW", data=df00z, kind="reg",
                   color="r", size=7)
plt.show(g)
```
