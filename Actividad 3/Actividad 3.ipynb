```python
#Importaremos las bibliotecas a utilizar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Leeremos los datos meteorológicos del 22 de Diciembre de Brindisi, Italia
dfDec0=pd.read_csv("BrindisiDec.txt", sep="\s+")
```

```python
#Leeremos los datos meteorológicos del 22 de Junio de Brindisi, Italia
dfJun0=pd.read_csv("BrindisiJun.txt", sep="\s+")
```

```python
dfDec1=pd.DataFrame(dfDec0) 
```

```python
dfJun1=pd.DataFrame(dfJun0)
```

```python
#Graficaremos presión contra altura en Diciembre
dfDec2=dfDec1[["HGHT", "PRES"]]
dfDec2.plot(x="PRES");  plt.legend(["PRES"])
plt.title("Variación de la presion con respecto a la altura (Diciembre)")
plt.xlabel("Presion(hPa)")
plt.ylabel("Altura(m)")
plt.xlim([0,1000])
plt.grid(True)
plt.show()
```

```python
#Graficaremos presión contra altura en Junio
dfJun2=dfJun1[["HGHT", "PRES"]]
dfJun2.plot(x="PRES", color="red"); plt.legend(["PRES"])
plt.title("Variación de la presion con respecto a la altura (Junio)")
plt.xlabel("Presión(hPa)")
plt.ylabel("Altura(m)")
plt.xlim([0,1000])
plt.grid(True)
plt.show()
```

```python
#Graficaremos temperatura contra altura en Diciembre
dfDec2=dfDec1[["HGHT", "TEMP"]]
dfDec2.plot(x="TEMP"); plt.legend(["TEMP"])
plt.title("Variación de la temperatura con respecto a la altura (Diciembre)")
plt.xlabel("Temperatura(°C)")
plt.ylabel("Altura(m)")
plt.xlim([-70, 10])
plt.grid(True)
plt.show()
```

```python
#Graficaremos temperatura contra altura en Junio
dfJun2=dfJun1[["HGHT", "TEMP"]]
dfJun2.plot(x="TEMP", color="red"); plt.legend(["TEMP"])
plt.title("Variación de la temperatura con respecto a la altura (Junio)")
plt.xlabel("Temperatura(°C)")
plt.ylabel("Altura(m)")
plt.xlim([-70, 25])
plt.grid(True)
plt.show()
```

```python
#Graficaremos la temperatura (TEMP) y temperatura de rocío (DEWPT) como función de la altura para el mes de Diciembre
plt.plot(dfDec1.TEMP, dfDec1.HGHT); 
plt.plot(dfDec1.DWPT, dfDec1.HGHT);  plt.legend(["TEMP", "DWPT"])

plt.title("Variación de la temperatura y temperatura de rocío con respecto a la altura (Diciembre)")
plt.xlabel("Temperatura/ Temperatura de rocío (°C)")
plt.ylabel("Altura(m)")
plt.xlim([-100, 15])
plt.grid(True)
plt.show()
```

```python
#Graficaremos la temperatura (TEMP) y temperatura de rocío (DEWPT) como función de la altura para el mes de Junio
plt.plot(dfJun1.TEMP, dfJun1.HGHT); 
plt.plot(dfJun1.DWPT, dfJun1.HGHT);  plt.legend(["TEMP", "DWPT"])

plt.title("Variación de la temperatura y temperatura de rocío con respecto a la altura (Junio)")
plt.xlabel("Temperatura/ Temperatura de rocío (°C)")
plt.ylabel("Altura(m)")
plt.xlim([-90,25 ])
plt.grid(True)
plt.show()
```

```python
#Graficaremos la rapidez de los vientos en nudos (SKNT) en función de la altura de Diciembre. 
#Los nudos equivalen a 1852km/hr
dfDec2=dfDec1[["HGHT", "SKNT"]]
dfDec2.plot(x="SKNT"); plt.legend(["SKNT"])
plt.title("Variación de la rapidez de los vientos con respecto a la altura (Diciembre)")
plt.xlabel("Rapidez de los vientos (nudos)")
plt.ylabel("Altura(m)")
plt.xlim([-10,200])
plt.grid(True)
plt.show()
```

```python
#Graficaremos la rapidez de los vientos en nudos (SKNT) en función de la altura de Junio. 
#Los nudos equivalen a 1852km/hr
dfJun2=dfJun1[["HGHT", "SKNT"]]
dfJun2.plot(x="SKNT", color="red"); plt.legend(["SKNT"])
plt.title("Variación de la rapidez de los vientos con respecto a la altura (Junio)")
plt.xlabel("Rapidez de los vientos (nudos)")
plt.ylabel("Altura(m)")
plt.xlim([0 ,30])
plt.grid(True)
plt.show()
```

```python
#la humedad relativa(RELH en %) en función de la altura de Diciembre

dfDec2=dfDec1[["HGHT", "RELH"]]
dfDec2.plot(x="RELH"); plt.legend(["RELH"])
plt.title("Variación de la humedad relativa con respecto a la altura (Diciembre)")
plt.xlabel("Humedad relativa (%)")
plt.ylabel("Altura(m)")
plt.xlim([0 ,70])
plt.grid(True)
plt.show()
```

```python
#la humedad relativa(RELH en %) en función de la altura de Junio

dfJun2=dfJun1[["HGHT", "RELH"]]
dfJun2.plot(x="RELH", color="red"); plt.legend(["RELH"])
plt.title("Variación de la humedad relativa con respecto a la altura (Junio)")
plt.xlabel("Humedad relativa (%)")
plt.ylabel("Altura(m)")
plt.xlim([0 ,70])
plt.grid(True)
plt.show()
```
